```python
import pandas as pd
import matplotlib.pyplot as plt
df = pd.read_csv(r"C:\Users\chait\Downloads\Dataset1 - Sheet2.csv")
df2 = pd.read_csv(r"C:\Users\chait\Downloads\Dataset2 - Sheet2.csv")
print("\n===== Original Employee Data =====")
print(df)
```

```python
subset = df[['Name', 'Department', 'Salary']]
print("\n===== Subsetting =====")
print(subset)
```

```python
filter_salary = df[df['Salary'] > 60000]
print("\n===== Filtering (Salary > 60000) =====")
print(filter_salary)
```

```python
sort_by_age = df.sort_values(by='Age')
print("\n===== Sorting by Age (ascending) =====")
print(sort_by_age)
```

```python
sort_by_salary_desc = df.sort_values(by='Salary', ascending=False)
print("\n===== Sorting by Salary (descending) =====")
print(sort_by_salary_desc)
```

```python
avg_salary_dept = df.groupby('Department')['Salary'].mean()
print("\n===== Aggregating (Average Salary by Department) =====")
print(avg_salary_dept)
```

```python
merged = pd.merge(df, df2, on="Department")
print("\n===== Merging employee & department data =====")
print(merged)
```

```python
gender_counts = df.groupby('Gender')['ID'].count()
print("\n===== Aggregating (Count of employees by Gender) =====")
print(gender_counts)
```

```python
df_indexed = df.set_index('Department')
df2_indexed = df2.set_index('Department')
joined = df_indexed.join(df2_indexed)
print("\n===== Joining on Department index =====")
print(joined)
```

```python
pivot_table = df.pivot_table(values='Salary', index='Department', columns='Gender', aggfunc='mean')
print("\n===== Pivot Table (Avg Salary: Dept vs Gender) =====")
print(pivot_table)
```
