```python
import re
def extract_phone_numbers():
    text = input("Enter text containing phone numbers: ")
    pattern = r'\+?\d{1,3}[- ]?\d{10}|\d{3}[- ]\d{8}|\d{10}'
    phones = re.findall(pattern, text)
    print(phones)
```

```python
def find_capital_words():
    text = input("Enter text: ")
    caps = re.findall(r'\b[A-Z]{2,}\b', text)
    print(caps)
```

```python
def replace_multiple_spaces():
    text = input("Enter text: ")
    cleaned = re.sub(r'\s+', ' ', text)
    print(cleaned)
```

```python
def check_password_strength():
    password = input("Enter password: ")
    pattern = r'^(?=.*[a-z])(?=.*[A-Z])(?=.*\d)(?=.*[@$!%*?&])[A-Za-z\d@$!%*?&]{8,}$'
    if re.match(pattern, password):
        print("Strong")
    else:
        print("Weak")
```

```python
while True:
    print("1. Extract phone numbers")
    print("2. Find capital words")
    print("3. Replace multiple spaces")
    print("4. Check password strength")
    print("5. Exit")

    choice = input("Enter choice: ")

    if choice == "1":
        extract_phone_numbers()
    elif choice == "2":
        find_capital_words()
    elif choice == "3":
        replace_multiple_spaces()
    elif choice == "4":
        check_password_strength()
    elif choice == "5":
        break
    else:
        print("Invalid choice")
```
